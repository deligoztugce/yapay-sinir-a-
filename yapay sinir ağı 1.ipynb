```python
import numpy as np

def sigmoid(x):
    # Sigmoid fonksiyonu: 1 / (1 + e^(-x))
    # türevi: -(1 + e^(-x - 1))^2
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    
    # Sigmoid'in türevi: s(x) * (1 - s(x))
    return sigmoid(x) * (1 - sigmoid(x))



# İleri yayılım fonksiyonları
def l1f1(x1, x2):
    return x1 * w1 + x2 * w3

def l1f2(x1, x2):
    return x1 * w2 + x2 * w4

def of1(x1, x2):
    return x1 * w5 + x2 * w6

# Girişler ve hedef değer
i1 = 0.5
i2 = 0.3
y = 1  # Hedef değer

# İleri yayılım
h1_input = l1f1(i1, i2)
h1_output = sigmoid(h1_input)

h2_input = l1f2(i1, i2)
h2_output = sigmoid(h2_input)

h3_input = of1(h1_output, h2_output)
h3_output = sigmoid(h3_input)

# Kayıp fonksiyonu (MSE)
loss = (h3_output - y) ** 2
print("Çıkış:", h3_output)
print("Kayıp:", loss)

# Geri yayılım
# Çıkış katmanı türevleri
dL_dh3 = 2 * (h3_output - y)  # Kayıp fonksiyonunun h3'e göre türevi
print("dL_dh3:", dL_dh3)
dh3_input = sigmoid_derivative(h3_input)  # Sigmoid türevi
dL_dh3_input = dL_dh3 * dh3_input  # Zincir kuralı

# Çıkış katmanı ağırlıklarının türevleri
dL_dw5 = dL_dh3_input * h1_output
dL_dw6 = dL_dh3_input * h2_output

# Çıkış katmanı gizli katman çıkışlarına etkisi
dL_dh1_output = dL_dh3_input * w5
dL_dh2_output = dL_dh3_input * w6

# Gizli katman türevleri
dh1_input = sigmoid_derivative(h1_input)
dh2_input = sigmoid_derivative(h2_input)

dL_dh1_input = dL_dh1_output * dh1_input
dL_dh2_input = dL_dh2_output * dh2_input

# Gizli katman ağırlıklarının türevleri
dL_dw1 = dL_dh1_input * i1
dL_dw3 = dL_dh1_input * i2

dL_dw2 = dL_dh2_input * i1
dL_dw4 = dL_dh2_input * i2

print("\nAğırlık türevleri:")
print("dL_dw1:", dL_dw1)
print("dL_dw2:", dL_dw2)
print("dL_dw3:", dL_dw3)
print("dL_dw4:", dL_dw4)
print("dL_dw5:", dL_dw5)
print("dL_dw6:", dL_dw6)

# Ağırlıkların güncellenmesi
learning_rate = 0.1
w1 -= learning_rate * dL_dw1
w2 -= learning_rate * dL_dw2
w3 -= learning_rate * dL_dw3
w4 -= learning_rate * dL_dw4
w5 -= learning_rate * dL_dw5
w6 -= learning_rate * dL_dw6

print("\nGüncellenmiş Ağırlıklar:")
print("w1:", w1)
print("w2:", w2)
print("w3:", w3)
print("w4:", w4)
print("w5:", w5)
print("w6:", w6)
```

```python
ogrenciler = [
    {
        "vize": 0.6,
        "final": 0.5,
        "sonuc": 1
    },
    {
        "vize": 0.16,
        "final": 0.5,
        "sonuc": 0
    },
    {
        "vize": 0.69,
        "final": 0.57,
        "sonuc": 1
    },
    {
        "vize": 0.86,
        "final": 0.95,
        "sonuc": 1
    },
    {
        "vize": 0.69,
        "final": 0.57,
        "sonuc": 1
    }
]

epochs = 101
learning_rate = 1

w1 = 0.7
w2 = 0.3
w3 = 0.4
w4 = 0.6
w5 = 0.55
w6 = 0.45

for epoch in range(epochs):
  toplam_kayip = 0
  for ogrenci in ogrenciler:
    vize = ogrenci["vize"]
    final = ogrenci["final"]
    sonuc = ogrenci["sonuc"]

    mustafa_hoca = l1f1(vize, final)
    ali_hoca = l1f2(vize, final)

    final_deger = of1(mustafa_hoca, ali_hoca)
    kayip = loss_or_cost_error(sonuc, final_deger)
    toplam_kayip += kayip

    dC_o1 = loss_or_cost_error_derivative(sonuc, final_deger)
    do1_sigmoid = sigmoid_derivative(final_deger)
    do1_w5 = mustafa_hoca
    do1_w6 = ali_hoca

    dC_w5 = dC_o1 * do1_sigmoid * do1_w5
    dC_w6 = dC_o1 * do1_sigmoid * do1_w6

    do1_mustafa_hoca = w5
    do1_ali_hoca = w6

    dmustafa_hoca_sigmoid = sigmoid_derivative(mustafa_hoca)
    dali_hoca_sigmoid = sigmoid_derivative(ali_hoca)

    do1_mustafa_hoca = dC_o1 * do1_sigmoid * do1_mustafa_hoca
    do1_ali_hoca = dC_o1 * do1_sigmoid * do1_ali_hoca

    dC_w1 = do1_mustafa_hoca * dmustafa_hoca_sigmoid * vize
    dC_w2 = do1_ali_hoca * dali_hoca_sigmoid * vize
    dC_w3 = do1_mustafa_hoca * dmustafa_hoca_sigmoid * final
    dC_w4 = do1_ali_hoca * dali_hoca_sigmoid * final

    w1 -= learning_rate * dC_w1
    w2 -= learning_rate * dC_w2
    w3 -= learning_rate * dC_w3
    w4 -= learning_rate * dC_w4
    w5 -= learning_rate * dC_w5
    w6 -= learning_rate * dC_w6

  ortalama_kayip = toplam_kayip / len(ogrenciler)
  if ortalama_kayip < 0.01:
    print("Epoch:", epoch + 1, "Ortalama Kayıp:", round(ortalama_kayip, 3))
    break

  if epoch % 100 == 0:
    print("Epoch:", epoch + 1, "Ortalama Kayıp:", round(ortalama_kayip, 3))
  toplam_kayip = 0
```

```python
import numpy as np

# Hyperparameters
epochs = 101
learning_rate = 1

# Input matrix: 5 examples with [vize, final] scores
matrix_i = np.array([
    [0.5, 0.3],  # Example 1
    [0.6, 0.8],  # Example 2
    [0.2, 0.1],  # Example 3
    [0.9, 0.7],  # Example 4
    [0.4, 0.5]   # Example 5
])

# True outputs (sonuc): corresponding to each example
y = np.array([
    [1],  # Example 1
    [0],  # Example 2
    [1],  # Example 3
    [0],  # Example 4
    [1]   # Example 5
])

# Layer weights initialization
matrix_l1w = np.array([[0.7, 0.3],  # w1, w2
                       [0.4, 0.6]])  # w3, w4

matrix_ow = np.array([[0.55],  # w5
                      [0.45]])  # w6

# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

# Training loop
for epoch in range(epochs):
    toplam_kayip = 0

    # Forward pass
    o1 = np.dot(matrix_i, matrix_l1w)  # First layer outputs
    so1 = sigmoid(o1)                 # Activation at first layer
    o = np.dot(so1, matrix_ow)        # Final output before activation
    so = sigmoid(o)                   # Final output after activation

    # Loss computation (Mean Squared Error)
    kayip = 0.5 * np.mean((y - so) ** 2)
    toplam_kayip += kayip

    # Backward pass
    # Derivatives w.r.t final output
    dC_o = so - y                               # Loss derivative
    do_sigmoid = sigmoid_derivative(o)          # Sigmoid derivative
    dC_matrix_ow = np.dot(so1.T, dC_o * do_sigmoid)  # Gradients for output weights

    # Derivatives w.r.t first layer
    dC_so1 = np.dot(dC_o * do_sigmoid, matrix_ow.T)  # Gradients for first layer activation
    do1_sigmoid = sigmoid_derivative(o1)            # Sigmoid derivative
    dC_matrix_l1w = np.dot(matrix_i.T, dC_so1 * do1_sigmoid)  # Gradients for first layer weights

    # Update weights
    matrix_ow -= learning_rate * dC_matrix_ow
    matrix_l1w -= learning_rate * dC_matrix_l1w

    # Compute average loss
    ortalama_kayip = toplam_kayip

    # Stop early if loss is small
    if ortalama_kayip < 0.01:
        print("Epoch:", epoch + 1, "Ortalama Kayıp:", round(ortalama_kayip, 3))
        break

    # Log progress every 100 epochs
    if epoch % 100 == 0:
        print("Epoch:", epoch + 1, "Ortalama Kayıp:", round(ortalama_kayip, 3))
```
